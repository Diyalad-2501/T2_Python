```python
d={'D':10,'B':9,'C':8}
print(len(d))
print(max(d))
print(min(d))
print(sorted(d))

for i in d:
    print(i)
for i in d.values():
    print(i)
for i,j in d.items():
    print(i,j)
```

```python
d={7:[7,3,2] , 6:[1,5,4],8:[3,2,1]}

for i,j in d.items():
    d[i]=sorted(j)

d=dict(sorted(d.items()))
print(d)
```

```python
l={'a': ['apple', 'ant', 'air', 'an', 'a'], 'm': ['mango', 'man'], 'r': ['right'], 'h': ['hi', 'hello']}

def f(x):
    return len(x[1])

l=dict(sorted(l.items(),key=f))
print(l)
```
