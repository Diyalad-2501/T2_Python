```python
num = int(input("Enter a number: "))
d={1:'One' , 2:'Two' , 3:'Three' , 4:'Four' , 5:'Five' , 6:'Six' , 7:'Seven' , 8:'Eight' , 9:'Nine' , 10:'Ten' , 11:'Eleven' ,12:'Twelve' ,13:'Thirteen' ,14:'Fourteen' ,15:'Fifteen',16:'Sixteen' ,17:'Seventeen',18:'Eighteen',19:'Nineteen' ,20:'Twenty' ,30:'Thirty',40:'Fourty' , 50:'Fifty' , 60:'Sixty',70:'Seventy' , 80:'Eighty',90:'Ninety'}

if num<21:
    print(d[num])
elif num<100:
    if num in d:
        print(d[num])
    else:
        print(d[(num//10)*10], d[num%10])
else:
    if num%100==0:
        print(d[num//100] , 'Hundred')
    else:
        print(d[num//100] ,'Hudred',end=" ")
        num=num%100      
        if num<21:
            print(d[num])
        elif num<100:
            if num in d:
                print(d[num])
            else:
                print(d[(num//10)*10], d[num%10])
```

```python
n = input("Enter roman number: ")
roman = {'I':1 , 'V':5 , 'X':10,'L':50 , 'C':100 , 'D':500 , 'M':1000}
sum1=0
for i in range(len(n)):
    if i+1<len(n) and roman[n[i]]<roman[n[i+1]]:
        sum1-=roman[n[i]]
    else:
        sum1+=roman[n[i]]
print(sum1)
```

```python
s='()()()'

if s[0]==']' or s[0]==')' or s[0]=='}':
    b=False
else:
    for i in range(len(s)//2):
        if (s[i]=='[' and s[-(i+1)]==']' )or( s[i]=='{' and s[-(i+1)]=='}' )or (s[i]=='(' and s[-(i+1)]==')'):
            b=True
        else:
            b=False
print('Balanced' if b==True else 'Not balanced')
```
