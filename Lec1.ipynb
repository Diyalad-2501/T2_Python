```python
s1 = input('Enter 1st string: ')
s2 = input('Enter 2nd string: ')
count=0
count1=0
for i in s1:
    for j in s2:
        if i==j:
            count+=1
        else:
            continue
for i in s2:
    for j in s1:
        if i==j:
            count1+=1
        else:
            continue
if count==count1 and len(s1)==len(s2):
    print('It is balanced')
else:
    print("It is not balanced")
```
