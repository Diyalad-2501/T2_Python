```python
s=input("Enter a string: ")
s=s.split(" ")
s=sorted(s)
print(s)
ans=""

if len(s[0])<len(s[-1]) :
    length = len(s[0])
else:
    length = len(s[-1])
    
for i in range(length):
    if s[0][i]==s[-1][i]:
        ans+=s[0][i]
    else:
        break
        
print(ans)
    
```

```python
bucket = eval(input("Enter bucket"))
ans=0
left = 0
right = len(bucket)-1
c_ans = 0
for i in range(len(bucket)):
        c_ans = min(bucket[left] , bucket[right]) *(right - left)
        ans= max(ans , c_ans)
        
        if bucket[left]<bucket[right]:
            left+=1
        else:
            right-=1
print(ans)
```

```python
bracket = '([{]})'
stack = []
d = {'}':'{' , ']':'[' , ')':'('}

def f(bracket):
    for i in bracket:
        if i in '[({':
            stack.append(i)
        else:
            if len(stack)==0:
                return False
            elif d[i]!=stack.pop():
                return False
    if len(stack)!=0:
        return False
    else:
        return True

print('Balanced' if f(bracket) else 'Not balanced')
```
