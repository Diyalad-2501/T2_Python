```python
password = input("Enter your password: ")
countup=0
countlow=0
countdigit=0
countspace=0
countsp=0

if len(password)<=8:
    print("Password length should be more than 8")
for i in password:    
    if i.isupper():
        countup+=1
    elif i.islower():
        countlow+=1
    elif i.isnumeric():
        countdigit+=1
    elif i.isspace():
        countspace+=1
    else:
        countsp+=1
        
if countup>1 and countlow>1 and countdigit>1 and countsp>1:
    print("Valid")
else:
    if countup<2:
        print("Atleast two upper case letter required")
    if countlow<2:
        print("Atleast two lower case letter required")
    if countdigit<2:
        print("Atleast two digit required")
    if countsp<2:
        print("Atleast two special character required")
    if countspace>0:
        print("No space allowed")
```

```python
message = input("Enter a message: ")
key = int(input("Enter key: "))
tochar=0

for i in message:
    if key==2:
        tochar=ord(i)
        if tochar>=65 and tochar<=90:
            if tochar==89:
                print(chr(65),end="")
            elif tochar==90:
                tochar=64
        else:
            if tochar>=97 and tochar<=122:
                if tochar==121:
                    print(chr(97),end="")
                elif tochar==122:
                    tochar=96
        print(chr(tochar+2),end="")
```

```python
s=input()
key=int(input())
Alpha='ABCDEFGHIJKLMNOPQRSTUVWXYZ'
alpha=Alpha.lower()
digit='0123456789'
new=""
for i in s:
    if i in Alpha:
        new+=Alpha[((Alpha.find(i)+key)%len(Alpha))]
    elif i in alpha:
        new+=alpha[((alpha.find(i)+key)%len(alpha))]
    elif i in digit:
        new+=digit[((digit.find(i)+key)%len(digit))]
    else:
        new = new+i
print(new)
```
