```python
L = eval(input("Enter a list"))
count=0

for i in range(len(L)):
    new = L[:]
    new.pop(i)
    
    if sum(new[::2])==sum(new[1::2]):
#         print(i)
        count+=1
        
print(count)
```

```python
n = eval(input('Enter a list: '))
count=0
even=0
odd=0

for i,j in enumerate(n):
    for r,s in enumerate(n[i+1::]):
        if r%2==0:
            even+=s
        else:
            odd+=s
        
        if even==odd:
            count+=1
            
print(count)
```

```python
L = eval(input("Enter a list: "))
sum1=0

for i in range(len(L)):
    if L[i]==13 or L[i-1]==13 and i!=0:
        continue
    else:
        sum1+=L[i]
print(sum1)
```
