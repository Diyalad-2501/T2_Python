```python
t = eval(input('Enter tuple:'))
even = tuple()
odd = tuple()

for i in t:
    if i%2==0:
        even=even+(i,)
    else:
        odd= odd+(i,)
print("Even tuple:",even)
print("Odd tuple:",odd)
```

```python
t = eval(input('Enter tuple: '))
even = tuple()
odd = tuple()

for i in t:
    if i%2==0:
        even+=(i,)
    else:
        odd+=(i,)
print("Max Even",max(even))
print("Max Odd",max(odd))
print("Min even",min(even))
print("Min odd",min(odd))

# x = tuple(i for i in t if i%2==0)
# y = tuple(i for i in t if i%2!=0)
```

```python
l = eval(input('Enter tuple:'))
candy=[1]*len(l)
print(candy)

for i in range(1,len(l)):
    if l[i]>l[i-1]:
        candy[i]=candy[i-1]+1
print(candy)
for i in range(-2,-len(l)-1,-1):
    if l[i]>l[i+1]:
        candy[i]=max(candy[i],candy[i+1]+1)
            
print(candy)
print(sum(candy))
```
